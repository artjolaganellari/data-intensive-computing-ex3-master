```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize': (15, 8)})
```

```python
# Local inference
inf_local_data = list()
with open('inference_all_local.txt', 'r') as inf_local:
    inf_local_data = np.array(eval(inf_local.read()))

print(f'Mean: {inf_local_data.mean()*1000}ms')
print(f'Total: {inf_local_data.sum()}s')
```

```python
# Remote inference
inf_remote_data = list()
with open('inference_times_remote.txt', 'r') as inf_remote:
    inf_remote_data = np.array(eval(inf_remote.read()))

print(f'Mean: {inf_remote_data.mean()*1000}ms')
print(f'Total: {inf_remote_data.sum()}s')
```

```python
upload_data = list()
with open('upload_times.txt', 'r') as upload_file:
    upload_data = np.array(eval(upload_file.read()))

print(f'Mean: {upload_data.mean()*1000}ms')
print(f'Total: {upload_data.sum()}s')
```

```python
sns.distplot(upload_data)
plt.legend(labels=['Transfer time'], fontsize='x-large')
```

```python
sns.distplot(inf_local_data[1:])
ax = sns.distplot(inf_remote_data[1:])
plt.legend(labels=['Local inference', 'Remote inference'], fontsize='x-large')
```
